```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Downloads/Uber Request Data (2).csv')
df.head()
```

```python
df.info()
```

```python
df['Request timestamp'] = pd.to_datetime(
    df['Request timestamp'],
    errors='coerce',
    dayfirst=True
)

df['Drop timestamp'] = pd.to_datetime(
    df['Drop timestamp'],
    errors='coerce',
    dayfirst=True
)
```

```python
df.isna().sum()
```

```python
df['Status'] = df['Status'].str.strip()
df['Status'].unique()
```

```python
df['Pickup point'] = df['Pickup point'].str.strip()
df['Pickup point'].unique()
```

```python
df.duplicated(subset='Request id').sum()
```

```python
df = df.drop_duplicates(subset='Request id')
```

```python
df['Hour_of_Day'] = df['Request timestamp'].dt.hour
```

```python
df['Day_of_Week'] = df['Request timestamp'].dt.day_name()
```

```python
# Completed trips should have Drop timestamp
df.loc[
    (df['Status'] == 'Trip Completed') & (df['Drop timestamp'].isna())
]
```

```python
df.info()
df.isna().sum()
```

```python
df.to_csv("Uber_Final_Cleaned_Data.csv", index=False)
```
